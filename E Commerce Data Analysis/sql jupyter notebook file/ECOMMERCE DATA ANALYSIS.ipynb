```python
1️⃣ Total KPIs
```

```python
query = """
SELECT 
    SUM(Sales) AS Total_Sales,
    SUM(Profit) AS Total_Profit,
    COUNT(DISTINCT [Order ID]) AS Total_Orders,
    COUNT(DISTINCT [Customer ID]) AS Total_Customers
FROM dbo.Orders$;
"""
pd.read_sql(query, conn)
```

```python
2️⃣ Total Sales by Month-Year
```

```python
query = """
SELECT 
    FORMAT([Order Date], 'yyyy-MM') AS MonthYear,
    SUM(Sales) AS Total_Sales
FROM dbo.Orders$
GROUP BY FORMAT([Order Date], 'yyyy-MM')
ORDER BY MonthYear;
"""
pd.read_sql(query, conn)
```

```python
3️⃣ Top 10 Products by Total Sales
```

```python
query = """
SELECT TOP 10 
    [Product Name],
    SUM(Sales) AS Total_Sales
FROM dbo.Orders$
GROUP BY [Product Name]
ORDER BY Total_Sales DESC;
"""
pd.read_sql(query, conn)
```

```python
4️⃣ Total Profit by Category and Sub-Category
```

```python
query = """
SELECT
    Category,
    [Sub-Category],
    SUM(Sales) AS Total_Sales,
    SUM(Profit) AS Total_Profit
FROM dbo.Orders$
GROUP BY Category, [Sub-Category]
ORDER BY Total_Profit DESC;
"""
pd.read_sql(query, conn)
```
